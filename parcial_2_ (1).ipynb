```python
import pandas as pd
import sqlite3
import os

class CafeteriaCTO:
    def __init__(self, db_name='parcial_2_cafeteria.db'):
        self.db_name = db_name
        self.conn = None

    def fase1_extraccion(self):
        """Carga los datos y genera los archivos CSV sucios."""
        # Datos proporcionados en el enunciado
        prod_data = """id_producto,producto_categoria,precio,stock,fecha_vencimiento
1,Café Tostao - Bebida,$5000,50,2026-12-01
2,chocolatina jet - snack,1200,NaN,2027-01-15
3,Empanada - Snack,$3000,20,2026-05-10
4,Jugo Hit - Bebida,2500,NaN,2026-08-22
5,Galletas Festival - Snack,$1500,100,2026-11-30
6,agua cristal - bebida,2000,30,2026-10-10
7,Sandwich de Pavo - ALMUERZO,$7500,15,2026-04-20
8,brownie - Postre,3500,NaN,2026-04-25
9,Té Helado - BEBIDA,$3000,40,2026-07-11
10,Papas Margarita - snack,2200,60,2026-09-05
11,Croissant - Panaderia,$2800,NaN,2026-05-15
12,Avena Alpina - Bebida,2600,25,2026-06-20
13,Yogurt Finesse - BEBIDA,$3200,35,2026-07-30
14,Muffin de Arandanos - postre,4000,NaN,2026-08-10
15,Dedito de Queso - Snack,$2500,45,2026-05-05
16,Ensalada de Frutas - Postre,5500,10,2026-04-18
17,Wrap de Pollo - Almuerzo,$8000,NaN,2026-04-19
18,Gomitas Trululu - SNACK,1000,80,2026-12-20
19,Tinto - Bebida,$1500,100,2026-04-30
20,Aromatica - bebida,2000,NaN,2026-05-01"""

        clie_data = """id_cliente,cliente_tipo,email,telefono,edad,fecha_nacimiento
101,Diego Zuluaga - Profesor,diego@correo.com,3001112233,40,1985-05-10
102,ana gomez - estudiante,,NaN,22,2004-01-15
103,CARLOS RUIZ - PROFESOR,carlos@correo.com,3109998877,35,1989-08-20
104,maria lopez - estudiante,,3204445566,28,1995-11-30
105,Luis Perez - Externo,luis@correo.com,NaN,50,1976-02-14
106,sofia ramirez - ESTUDIANTE,sofia@correo.com,3112223344,20,2006-03-22
107,JUAN CAMILO - estudiante,juan@correo.com,NaN,24,2002-07-11
108,andres felipe - Externo,,3156667788,30,1996-09-05
109,VALERIA MORALES - profesor,valeria@correo.com,3009991122,45,1981-12-01
110,isabella carranza - ESTUDIANTE,isa@correo.com,NaN,21,2005-04-18
111,matias mondragon - estudiante,,3105556677,23,2003-02-28
112,ARIANNE AMOROCHO - externo,arianne@correo.com,3201112233,27,1999-06-15
113,mariana peña - Estudiante,mariana@correo.com,NaN,22,2004-08-08
114,nicolas torres - PROFESOR,nico@correo.com,3114445566,38,1988-10-10
115,juan leyton - estudiante,,3007778899,25,2001-11-25
116,CRISTIAN BERMUDEZ - externo,cristian@correo.com,NaN,32,1994-01-05
117,paula arciniegas - ESTUDIANTE,paula@correo.com,3152223344,20,2006-09-14
118,samuel lozano - estudiante,,3206667788,24,2002-05-30
119,GABRIELA RUIZ - profesor,gabi@correo.com,NaN,42,1984-07-07
120,lucas blanco - Externo,lucas@correo.com,3103334455,29,1995-03-12"""

        prov_data = """nit_proveedor,empresa_ciudad,contacto,telefono,email
900111,CoopCafe - Bogota,Carlos Perez,3005551122,carlos@coop.com
800222,Insumos Panaderos - Medellin,NaN,3114445566,
700333,Lacteos El Prado - Cali,Ana Rojas,NaN,ana@prado.com
600444,Distribuidora Sabana - Chia,Luis Gomez,3208889900,luis@sabana.com
500555,Salsas y Especias - Bogota,NaN,3157778899,
400666,Empaques de Carton - Barranquilla,Marta Diaz,3001112233,marta@empaques.com
300777,Granos y Cereales - Bucaramanga,NaN,NaN,ventas@granos.com
200888,Bebidas Refrescantes - Bogota,Jorge Silva,3109998877,
100999,Dulces y Postres - Medellin,Elena Castro,3204445566,elena@dulces.com
999000,Frutas Frescas - Cali,NaN,3112223344,frutas@cali.com"""

        # Guardado con codificación correcta
        with open('parcial_2_productos_sucios.csv', 'w', encoding='utf-8-sig') as f: f.write(prod_data)
        with open('parcial_2_clientes_sucios.csv', 'w', encoding='utf-8-sig') as f: f.write(clie_data)
        with open('parcial_2_proveedores_sucios.csv', 'w', encoding='utf-8-sig') as f: f.write(prov_data)

        return (pd.read_csv('parcial_2_productos_sucios.csv', encoding='utf-8-sig'),
                pd.read_csv('parcial_2_clientes_sucios.csv', encoding='utf-8-sig'),
                pd.read_csv('parcial_2_proveedores_sucios.csv', encoding='utf-8-sig'))

    def fase2_limpieza(self, df_p, df_c, df_pr):
        """Aplica normalización 1NF, 2NF, 3NF y limpieza profunda."""
        # PRODUCTOS
        df_p['stock'] = df_p['stock'].fillna(0).astype(int)
        df_p['precio'] = df_p['precio'].astype(str).str.replace('$', '', regex=False).astype(float)
        df_p[['nombre_producto', 'categoria']] = df_p['producto_categoria'].str.split(' - ', expand=True)
        df_p['nombre_producto'] = df_p['nombre_producto'].str.title()
        df_p.drop(columns=['producto_categoria'], inplace=True)

        # CLIENTES
        df_c['email'] = df_c['email'].fillna('No Registra').replace('', 'No Registra')
        df_c['telefono'] = df_c['telefono'].fillna('No Registra')
        df_c[['nombre_cliente', 'tipo_cliente']] = df_c['cliente_tipo'].str.split(' - ', expand=True)
        df_c['nombre_cliente'] = df_c['nombre_cliente'].str.title()
        df_c.drop(columns=['cliente_tipo', 'edad'], inplace=True) # 3NF: Eliminar edad redundante

        # PROVEEDORES
        df_pr[['empresa', 'ciudad']] = df_pr['empresa_ciudad'].str.split(' - ', expand=True)
        df_pr['contacto'] = df_pr['contacto'].fillna('Sin Contacto')
        df_pr['email'] = df_pr['email'].fillna('No Registra').replace('', 'No Registra')
        df_pr.drop(columns=['empresa_ciudad'], inplace=True)

        # Exportar Limpios
        df_p.to_csv('parcial_2_productos_limpios.csv', index=False, encoding='utf-8-sig')
        df_c.to_csv('parcial_2_clientes_limpios.csv', index=False, encoding='utf-8-sig')
        df_pr.to_csv('parcial_2_proveedores_limpios.csv', index=False, encoding='utf-8-sig')
        
        return df_p, df_c, df_pr

    def fase3_y_4_db_crud(self, df_p, df_c, df_pr):
        """Migración a SQLite y operaciones CRUD con SQL puro."""
        self.conn = sqlite3.connect(self.db_name)
        cursor = self.conn.cursor()
        cursor.execute("PRAGMA foreign_keys = ON;")

        # Re-crear tablas para asegurar integridad referencial
        cursor.execute("DROP TABLE IF EXISTS ventas")
        cursor.execute("DROP TABLE IF EXISTS productos")
        cursor.execute("DROP TABLE IF EXISTS clientes")
        cursor.execute("DROP TABLE IF EXISTS proveedores")

        cursor.execute("CREATE TABLE productos (id_producto INTEGER PRIMARY KEY, precio REAL, stock INTEGER, fecha_vencimiento TEXT, nombre_producto TEXT, categoria TEXT)")
        cursor.execute("CREATE TABLE clientes (id_cliente INTEGER PRIMARY KEY, email TEXT, telefono TEXT, fecha_nacimiento TEXT, nombre_cliente TEXT, tipo_cliente TEXT)")
        cursor.execute("CREATE TABLE proveedores (nit_proveedor INTEGER PRIMARY KEY, contacto TEXT, telefono TEXT, email TEXT, empresa TEXT, ciudad TEXT)")

        # Cargar datos limpios
        df_p.to_sql('productos', self.conn, if_exists='append', index=False)
        df_c.to_sql('clientes', self.conn, if_exists='append', index=False)
        df_pr.to_sql('proveedores', self.conn, if_exists='append', index=False)

        # Crear tabla VENTAS (Estructura exacta del parcial)
        cursor.execute("""
        CREATE TABLE ventas (
            id_venta INTEGER PRIMARY KEY AUTOINCREMENT,
            id_cliente INTEGER,
            id_producto INTEGER,
            id_proveedor INTEGER,
            cantidad INTEGER,
            total_venta REAL,
            fecha_venta TEXT,
            FOREIGN KEY (id_cliente) REFERENCES clientes(id_cliente),
            FOREIGN KEY (id_producto) REFERENCES productos(id_producto),
            FOREIGN KEY (id_proveedor) REFERENCES proveedores(nit_proveedor)
        )
        """)

        # INSERT: 5 registros
        ventas_list = [
            (101, 1, 900111, 2, 10000.0, '2026-04-15'),
            (102, 5, 500555, 3, 4500.0, '2026-04-15'),
            (105, 10, 200888, 1, 2200.0, '2026-04-15'),
            (110, 15, 400666, 2, 5000.0, '2026-04-15'),
            (120, 20, 999000, 5, 10000.0, '2026-04-15')
        ]
        cursor.executemany("INSERT INTO ventas (id_cliente, id_producto, id_proveedor, cantidad, total_venta, fecha_venta) VALUES (?,?,?,?,?,?)", ventas_list)

        # UPDATE: Venta 1
        cursor.execute("UPDATE ventas SET cantidad = 10, total_venta = 50000 WHERE id_venta = 1")

        # DELETE: Venta 3
        cursor.execute("DELETE FROM ventas WHERE id_venta = 3")

        self.conn.commit()

        # READ con JOIN
        print("\n--- REPORTE FINAL DE VENTAS ---")
        reporte = pd.read_sql("""
            SELECT c.nombre_cliente, p.nombre_producto, v.total_venta 
            FROM ventas v
            JOIN clientes c ON v.id_cliente = c.id_cliente
            JOIN productos p ON v.id_producto = p.id_producto
        """, self.conn)
        print(reporte)
        self.conn.close()

# Ejecución del Proyecto
if __name__ == "__main__":
    cto = CafeteriaCTO()
    df_p_sucio, df_c_sucio, df_pr_sucio = cto.fase1_extraccion()
    df_p, df_c, df_pr = cto.fase2_limpieza(df_p_sucio, df_c_sucio, df_pr_sucio)
    cto.fase3_y_4_db_crud(df_p, df_c, df_pr)
    print("\n✅ Parcial finalizado con arquitectura POO (Bono Incluido).")
```
